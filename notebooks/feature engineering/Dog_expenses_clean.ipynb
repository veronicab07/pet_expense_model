```python
# Create a dictionary mapping breed to tier
breed_to_tier = {
    'affenpinscher': 1, 'australian terrier': 1, 'basenji': 1, 'bedlington terrier': 1, 'border terrier': 1,
    'boston terrier': 1, 'cairn terrier': 1, 'cesky terrier': 1, 'chihuahua': 1, 'chinese crested': 1,
    'dachshund': 1, 'dandie dinmont terrier': 1, 'english toy spaniel': 1, 'english toy terrier': 1, 
    'glen of imaal terrier': 1, 'havanese': 1, 'italian greyhound': 1, 'japanese chin': 1, 'maltese': 1, 
    'manchester terrier': 1, 'miniature schnauzer': 1, 'norfolk terrier': 1, 'papillon': 1, 'pekingese': 1, 
    'scottish terrier': 1, 'sealyham terrier': 1, 'shetland sheepdog': 1, 'shih tzu': 1, 'silky terrier': 1, 
    'skye terrier': 1, 'smooth fox terrier': 1, 'west highland white terrier': 1, 'wire fox terrier': 1, 
    'yorkshire terrier': 1, 'australian shepherd': 1, 'basset hound': 1, 'border collie': 1, 'cardigan welsh corgi': 1,
    'norwegian buhund': 1, 'pembroke welsh corgi': 1, 'swedish vallhund': 1, 'spinone italiano': 1, 'sussex spaniel': 1,
    'small mix': 1, 'medium mix': 1, 'brussels griffon': 2, 'cavalier king charles spaniel': 2, 'kerry blue terrier': 2, 
    'king charles spaniel': 2, 'miniature bull terrier': 2, 'petit basset griffon vendeen': 2, 'pomeranian': 2, 
    'australian cattle dog': 2, 'beagle': 2, 'brittany': 2, 'cocker spaniel': 2, 'finnish lapphund': 2, 
    'german pinscher': 2, 'irish terrier': 2, 'polish lowland sheepdog': 2, 'siberian husky': 2, 
    'soft coated wheaten terrier': 2, 'whippet': 2, 'bearded collie': 2, 'field spaniel': 2, 'afghan hound': 2, 
    'alaskan malamute': 2, 'borzoi': 2, 'bouvier des flandres': 2, 'briard': 2, 'canaan dog': 2, 'clumber spaniel': 2, 
    'curly coated retriever': 2, 'english setter': 2, 'english springer spaniel': 2, 'german shorthaired pointer': 2, 
    'german wirehaired pointer': 2, 'giant schnauzer': 2, 'gordon setter': 2, 'greyhound': 2, 'irish red and white setter': 2, 
    'irish setter': 2, 'irish water spaniel': 2, 'komondor': 2, 'nova scotia duck tolling retriever': 2, 'otterhound': 2, 
    'pharaoh hound': 2, 'pointer': 2, 'portuguese water dog': 2, 'saluki': 2, 'samoyed': 2, 'vizsla': 2, 
    'welsh springer spaniel': 2, 'wirehaired pointing griffon': 2, 'large mix': 2, 'pug': 3, 'bull terrier': 3, 
    'french bulldog': 3, 'staffordshire bull terrier': 3, 'airedale terrier': 3, 'akita': 3, 'anatolian shepherd dog': 3, 
    'belgian tervuren': 3, 'bernese mountain dog': 3, 'bloodhound': 3, 'bullmastiff': 3, 'chesapeake bay retriever': 3, 
    'chow chow': 3, 'flat coated retriever': 3, 'german shepherd': 3, 'golden retriever': 3, 'great dane': 3, 
    'great pyrenees': 3, 'greater swiss mountain dog': 3, 'irish wolfhound': 3, 'labrador': 3, 'leonberger': 3, 
    'mastiff': 3, 'newfoundland': 3, 'old english sheepdog': 3, 'saint bernard': 3, 'scottish deerhound': 3, 
    'tibetan mastiff': 3, 'boxer': 4, 'dobermann': 4, 'rhodesian ridgeback': 4, 'rottweiler': 4, 'mix small': 1, 'mix medium': 1, 
    'mix large': 2, 'american english coonhound': 2, 'italian spinone': 2
}

# Map the dict to the breed column
df['tier'] = df['Breed'].map(breed_to_tier)

# Add the health costs based on tier
tier_costs = {
    1: {'Cost_1_5': 64 * 12, 'Cost_6_7': 74 * 12, 'Cost_8_plus': 136 * 12},
    2: {'Cost_1_5': 72 * 12, 'Cost_6_7': 83 * 12, 'Cost_8_plus': 152 * 12},
    3: {'Cost_1_5': 90 * 12, 'Cost_6_7': 105 * 12, 'Cost_8_plus': 194 * 12},
    4: {'Cost_1_5': 103 * 12, 'Cost_6_7': 119 * 12, 'Cost_8_plus': 222 * 12}
}

df['Cost_1_5'] = df['tier'].map(lambda x: tier_costs[x]['Cost_1_5'] if pd.notna(x) else 0)
df['Cost_6_7'] = df['tier'].map(lambda x: tier_costs[x]['Cost_6_7'] if pd.notna(x) else 0)
df['Cost_8_plus'] = df['tier'].map(lambda x: tier_costs[x]['Cost_8_plus'] if pd.notna(x) else 0)

# Calculate the total health cost based on longevity
def calculate_total_cost(row):
    cost_1_5_years = row['Cost_1_5'] * min(row['longevity'], 5)
    cost_6_7_years = row['Cost_6_7'] * min(max(row['longevity'] - 5, 0), 2)
    cost_8_plus_years = row['Cost_8_plus'] * max(row['longevity'] - 7, 0)
    return cost_1_5_years + cost_6_7_years + cost_8_plus_years

df['health_insurance_cost'] = df.apply(calculate_total_cost, axis=1)
```

```python
#create liability_insurance column
def liability_insurance(breed):
    if breed in ['bull terrier', 'staffordshire bull terrier']:
        return 20
    else:
        return 7

df['liability_insurance'] = df['Breed'].apply(liability_insurance)
```

```python
#create dog_tax column
def dog_tax(breed):
    if breed in ['bull terrier', 'staffordshire bull terrier']:
        return 50
    else:
        return 10

df['dog_tax'] = df['Breed'].apply(dog_tax)
```

```python
# Severity mapping 
severity_map = {
    'hips': 5, 'kidney': 5, 'bloat': 5, 'heart': 5,
    'elbows': 4, 'eyes': 4, 'spine': 4, 'liver': 4,
    'neurological': 4, 'thyroid': 3, 'allergies': 3,
    'osteopathy': 3, 'bladder': 3, 'patella': 2,
    'breathing': 2, 'bleeding': 1, 'cleft palate': 1,
    'dental': 1
}

# Function to calculate severity score based on genetic_ailments
def calculate_severity(ailments):
    total_severity = 0
    ailments_list = [ailment.strip() for ailment in ailments.split(',')]
    for ailment in ailments_list:
        if ailment in severity_map:
            total_severity += severity_map[ailment]
    return total_severity

# Apply severity calculation to create a new column 'severity_score'
df['severity_score'] = df['genetic_ailments'].apply(calculate_severity)
```
